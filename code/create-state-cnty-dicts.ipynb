```python
import pandas as pd
from datetime import date
import json

pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## Create select list state/county dicts

```python
cntys = pd.read_csv('../data/tl_2019_us_county.csv',dtype={'STATEFP':str,'COUNTYFP':str,'GEOID':str})
cntys = cntys[['STATEFP','COUNTYFP','GEOID','NAMELSAD']]

states = pd.read_csv('../data/us-state-fips.csv',dtype={'STATEFP':str})
states = states[['STATEFP','NAME']]
```

```python
state_fips_dict = dict(zip(states.STATEFP, states.NAME))     

with open('../app/assets/data/state-fips.json', 'w') as outfile:
    json.dump(state_fips_dict, outfile)
    
state_fips_dict
```

```python
cntys.head()
```

```python
cnty_group = cntys.groupby('STATEFP')
cnty_dict = {}

for key, selection in cnty_group:
    cnty_dict[key] = dict(zip(selection['GEOID'], selection['NAMELSAD']))

with open('../app/assets/data/cnty-fips.json', 'w') as outfile:
    json.dump(cnty_dict, outfile)
```
